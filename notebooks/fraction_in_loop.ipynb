### Based on `dist_loop`

```python
for cell_loop in (cells_loops1, cells_loops2, cells_loops3, cells_loops4):
    cell_loop['dist_loop'][cell_loop.beta == 1] = np.where(
        cell_loop[cell_loop.beta == 1]['in_loop'] > 0, cell_loop[cell_loop.beta == 1]['dist_loop'], 0)
for cell_loop in (cells_loops1, cells_loops2, cells_loops3, cells_loops4):
    dist_beta = np.array(
        [cell[1].iloc[0:5].dist_loop
            for cell in cell_loop[cell_loop.beta == 1].groupby('particle')])
    dist_nonbeta = np.array(
        [cell[1].iloc[0:5].dist_loop
            for cell in cell_loop[cell_loop.beta == 0].groupby('particle')])
    print((dist_beta.mean(axis=1).mean(), dist_nonbeta.mean(axis=1).mean()))
    print()
```

### Plotting

```python
cells_loops = (cells_loops1, cells_loops2, cells_loops3, cells_loops4)
plt.figure(figsize=(20,20))
labels = ('beta', 'nonbeta')
for i, cell_loop in enumerate(cells_loops):
    plt.subplot(4, 1, i + 1)
    plt.title('Cell intensities\nas function of time \n (all cells)', fontsize=20)
    plt.xlabel('Time index', fontsize=20)
    plt.ylabel('Intensity (from 0 to 1)', fontsize=20)
    labels = ('beta', 'nonbeta')
    for idx, cell_traj in cell_loop.groupby('particle'):
        beta = cell_traj.beta.values[0]
        x = cell_traj.frame.values
        y = cell_traj.intensity.values
        if np.random.rand() > 0.0:
            if beta:
                handle_beta = plt.plot(x, y, color='red', label='beta')
            else:
                handle_nonbeta = plt.plot(x, y, color='black', label='nonbeta')

    plt.legend(handles=[handle_beta[0], handle_nonbeta[0]], labels=labels, fontsize=30)
    
plt.tight_layout()
plt.show()
```
