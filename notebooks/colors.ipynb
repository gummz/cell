```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = np.zeros((512, 512, 4), dtype=np.float32)
image[:, :, (0, 2, 3)] = 1
plt.axis('off'), plt.imshow(image);
plt.imsave('color.png', image)
```

```python
color = plt.imread('color.png')
print(color.shape)
# color[:, :, 0]

plt.imshow(color), plt.axis('off');
```

```python
image = np.random.randint(low=0, high=255, size=(512, 512, 3))
plt.figure(figsize=(25, 30))
for i in range(3):
    plt.subplot(3, 1, i + 1)
    plt.imshow(image[:, :, i]);
```
