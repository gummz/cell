```python
def plot_calibration(calib_train, calib_val):
  fontsize = 20
  plt.figure(figsize=(15,10));
  plt.subplot(211)
  plt.bar(x, calib_train)
  plt.ylabel('Positive ratio', fontsize=fontsize)
  plt.xlabel('Certainty bins', fontsize=fontsize)
  plt.xticks(fontsize=fontsize, rotation=20)
  plt.yticks(fontsize=fontsize)
  plt.title('Model calibration (training)', fontsize=fontsize+5)
  plt.ylim(bottom=0, top=1)

  plt.subplot(212)
  plt.bar(x, calib_val)
  plt.ylabel('Positive ratio', fontsize=fontsize)
  plt.xlabel('Certainty bins', fontsize=fontsize)
  plt.xticks(fontsize=fontsize, rotation=20)
  plt.yticks(fontsize=fontsize)
  plt.title('Model calibration (validation)', fontsize=fontsize+5)
  plt.ylim(bottom=0, top=1)

  plt.tight_layout()
  plt.show()
```
