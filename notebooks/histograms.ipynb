### Imports

```python
import skimage
import matplotlib.pyplot as plt
import numpy as np
```

### Load brain data

```python
brain = skimage.data.brain()
```

```python
idx = 3
img = brain[idx]
img_mod = skimage.exposure.equalize_adapthist(img,
 kernel_size=10, clip_limit=1)
plt.figure(figsize=(15, 8))
plt.subplot(121)
plt.imshow(img), plt.axis('off')
plt.subplot(122)
plt.imshow(img_mod), plt.axis('off')
plt.show()
```

```python
img_hist = skimage.exposure.histogram(img)
mod_hist = skimage.exposure.histogram(img_mod)
plt.figure(figsize=(15, 8))
plt.subplot(121)
plt.hist(img)
plt.subplot(122)
plt.hist(img_mod)
plt.show()
```
